```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from PIL import Image
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from wordcloud import WordCloud
from nltk.tokenize import word_tokenize
# !pip install nltk
# !pip install wordcloud
```

```python
nltk.download('stopwords')
```

```python
stopwordlt = stopwords.words("english")
```

```python
file = open("./Dataset/Agatha.txt",'r').read()
```

```python
# file
```

```python
len(file)
```

```python
# # Download the punkt tokenizer model
# nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# #tonkenize word
words = word_tokenize(file)
```

```python
len(words)
```

```python
#Punctuation removal
word_no_punc = []
for word in words:
    if(word.isalpha()):
        word_no_punc.append(word.lower())
len(word_no_punc)
```

```python
# word_no_punc
```

```python
#Stopwords removal
clean_words = []
for  word in word_no_punc:
    if word not in stopwordlt:
        clean_words.append(word)
len(clean_words)
```

```python
# clean_words
```

```python
#join all the clean words to  form wordcloud
word_string = " ".join(clean_words)
# word_string
```

```python
#Generate wordcloud using WordCloud Method
wc = WordCloud(background_color = "white").generate(word_string)
plt.axis("off")
plt.imshow(wc)
```

```python
masking = np.array(Image.open("hack.jpg"))
plt.axis("off")
wc = WordCloud(mask=masking,background_color="white", contour_width=1,contour_color='black').generate(word_string)
plt.imshow(wc)
```
